```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np

```

```python
dizi_1= np.random.randint(20,80,5)
plt.plot(dizi_1)
plt.show()
```

```python
dizi_1= np.random.randint(20,80,5)
dizi_2= np.random.randint(20,80,5)
plt.plot(dizi_1)
plt.plot(dizi_2)
plt.show()
```

```python
dizi_1= np.random.randint(20,80,5)
dizi_2= np.random.randint(20,80,5)
uzunluk= np.arange(dizi_1.shape[0])
plt.plot(dizi_1)
plt.plot(dizi_2)

plt.xticks(uzunluk,["Deger 1","Deger 2","Deger 3","Deger 4","Deger 5"])
plt.show()
```

```python
dizi_1= np.random.randint(20,80,5)
dizi_2= np.random.randint(20,80,5)
uzunluk= np.arange(dizi_1.shape[0])

plt.plot(dizi_1,label="Dizi 1")
plt.plot(dizi_2,label="Dizi 2")
plt.legend()
plt.xticks(uzunluk,["Deger 1","Deger 2","Deger 3","Deger 4","Deger 5"])
plt.show()
```

```python
dizi_1= np.random.randint(20,80,5)
dizi_2= np.random.randint(20,80,5)
uzunluk= np.arange(dizi_1.shape[0])

plt.plot(dizi_1,label="Dizi 1")
plt.plot(dizi_2,label="Dizi 2")
plt.xlabel("dönem bilgisi")
plt.xlabel("not bilgisi")
plt.title("dönemlere göre notlar")
plt.legend()
plt.xticks(uzunluk,["Deger 1","Deger 2","Deger 3","Deger 4","Deger 5"])
plt.show()
```

```python
#12 aylık bir borsa verisini temsil eden grafiği oluşturun
kripto_1= np.random.randint(1000,50000,12)
kripto_2= np.random.randint(1000,50000,12)
uzunluk= np.arange(kripto_1.shape[0])

plt.plot(kripto_1,label="kripto 1")
plt.plot(kripto_2,label="kripto 2")
plt.xlabel("aylar")
plt.xlabel("fiyat/tl")
plt.title("2024 yılı kripto para fiyatları")
plt.legend()
plt.xticks(uzunluk,["ocak","şubat","mart","nisan","mayıs","haziran","temmuz","ağustos","eylül","ekim","kasım","aralık"],rotation=45)
plt.show()
```
