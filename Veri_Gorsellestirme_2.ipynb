```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Öğrenci sayısı
ogrenci_sayisi = 30

# Şube 1'in ders notları
s1_ders1_notlar = [75, 80, 85, 90]
s1_ders2_notlar = [70, 82, 88, 95]
s1_ders3_notlar = [65, 78, 80, 92]
s1_ders4_notlar = [72, 85, 90, 88]

# Şube 2'nin ders notları
s2_ders1_notlar = [78, 83, 88, 91]
s2_ders2_notlar = [68, 79, 84, 89]
s2_ders3_notlar = [71, 80, 86, 90]
s2_ders4_notlar = [75, 82, 87, 92]

# Ortalama hesaplama
s1_ders1_ortalama = sum(s1_ders1_notlar) / ogrenci_sayisi
s1_ders2_ortalama = sum(s1_ders2_notlar) / ogrenci_sayisi
s1_ders3_ortalama = sum(s1_ders3_notlar) / ogrenci_sayisi
s1_ders4_ortalama = sum(s1_ders4_notlar) / ogrenci_sayisi

s2_ders1_ortalama = sum(s2_ders1_notlar) / ogrenci_sayisi
s2_ders2_ortalama = sum(s2_ders2_notlar) / ogrenci_sayisi
s2_ders3_ortalama = sum(s2_ders3_notlar) / ogrenci_sayisi
s2_ders4_ortalama = sum(s2_ders4_notlar) / ogrenci_sayisi

# Bar grafikleri için verileri oluşturma
ders_ortalama_s1 = [s1_ders1_ortalama, s1_ders2_ortalama, s1_ders3_ortalama, s1_ders4_ortalama]
ders_ortalama_s2 = [s2_ders1_ortalama, s2_ders2_ortalama, s2_ders3_ortalama, s2_ders4_ortalama]
dersler = ['Ders 1', 'Ders 2', 'Ders 3', 'Ders 4']

x = np.arange(len(dersler))  # x ekseninde derslerin konumunu belirleme
bar_genisligi = 0.35

# Bar grafiklerini çizme
plt.bar(x, ders_ortalama_s1, width=bar_genisligi, label='Şube 1')
plt.bar(x + bar_genisligi, ders_ortalama_s2, width=bar_genisligi, label='Şube 2')
plt.xlabel('Dersler')
plt.ylabel('Ortalama Notlar')
plt.title('Her Bir Ders İçin Şube 1 ve Şube 2 Ortalama Notları')
plt.xticks(x + bar_genisligi/2, dersler)
plt.legend()

# Grafiği gösterme
plt.show()
```

```python
np.random.seed(0)  # Rastgele sayı üretimi için sabit bir başlangıç noktası belirleyelim

# 4 sınıfın 4 sınavdaki notlarını oluşturalım
notlar = np.random.randint(0, 101, size=(4, 4, 30))  # 4 sınıf, 4 sınav, 30 öğrenci

# Sınav notlarının ortalamalarını alalım
sınav_ortalamaları = np.mean(notlar, axis=2)  # Her bir sınavın ortalaması için son eksen boyunca ortalama alınır

# Her bir sınavın bar grafiğini çizdirme
fig, ax = plt.subplots(1, 4, figsize=(20, 5))  # 1 satır, 4 sütunlu bir subplot oluşturalım
sınavlar = ["Sınav 1", "Sınav 2", "Sınav 3", "Sınav 4"]

for i in range(4):
    ax[i].bar(np.arange(4), sınav_ortalamaları[i], color='skyblue')  # Her bir sınavın ortalamasını çizdirme
    ax[i].set_xticks(np.arange(4))  # x ekseni için etiketleri belirleme
    ax[i].set_xticklabels(["Şube 1", "Şube 2", "Şube 3", "Şube 4"])  # x ekseni etiketlerini belirleme
    ax[i].set_title(sınavlar[i])  # Grafik başlığını belirleme

plt.tight_layout()  # Grafiği sıkıştırma
plt.show()  # Grafiği gösterme
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Verileri oluşturma
notlar = np.random.randint(0, 100, (30, 4))  # 4 sınıfın 30 öğrencisinin notları

# Her sınıfın ortalamasını hesapla
ortalama_notlar = np.mean(notlar, axis=0)

# Sınıf etiketleri
siniflar = ['A', 'B', 'C', 'D']

# Renkler
renkler = ['blue', 'green', 'red', 'orange']

# Bar plotları yan yana çizme
bars = plt.bar(siniflar, ortalama_notlar, color=renkler)

plt.xlabel('Sınıflar')
plt.ylabel('Ortalama Notlar')
plt.title('Sınıf Ortalama Notları')

# Legend (Leyenda)
plt.legend(bars, siniflar)

for i in range(len(siniflar)):
    plt.text(i,ortalama_notlar[i],f"{ortalama_notlar[i]:.2f}",ha="center",va="bottom")
plt.show()
```
